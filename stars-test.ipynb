```python
import requests
import os
import json
from dotenv import load_dotenv
```

```python
load_dotenv()
```

```python
url = 'https://api.github.com/graphql'
```

```python
headers = {'Authorization': f'Bearer {os.getenv("GITHUB_TOKEN")}'}
```

```python
json_query = {
    "query": """query {
  repository(owner: "coinse", name: "sadl") {
    stargazers(first: 100) {
      edges {
        starredAt
        node {
          login
        }
      }
    }
  }
}"""
}
```

```python
print(json.dumps(requests.post(url, json=json_query, headers=headers).json(), indent=4))
```
